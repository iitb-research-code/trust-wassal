```python
import os
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
from ipywidgets import widgets, HBox, VBox
from IPython.display import display
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import pandas as pd

import os
import pandas as pd

def parse_stats_file(file_path):
    stats = {}
    with open(file_path, 'r') as file:
        for line in file:
            key, value = line.strip().split(": ")
            stats[key] = float(value.replace('%', ''))  # Convert to float, remove '%' if present
    return stats

def collect_data(base_path):
    data_records = []
    for root, dirs, files in os.walk(base_path):
        if 'simplex_data_stat.txt' in files:
            parts = root.split(os.sep)
            try:
                # Extracting details from the file path
                al_strategy = parts[-7]  # 8th from the end
                budget = parts[-6]       # 7th from the end
                experiment = parts[-5]   # 6th from the end
                round_num = parts[-1]    # 2nd from the end

                file_path = os.path.join(root, 'simplex_data_stat.txt')
                stats = parse_stats_file(file_path)
                correct = stats.get('Total Correct hypothesized labels (non-zero)', 0)
                incorrect = stats.get('Total Wrong hypothesized labels (non-zero)', 0)
                total = correct + incorrect
                accuracy = (correct / total) * 100 if total > 0 else 0
                for stat_name, stat_value in stats.items():
                    record = {
                        'AL Strategy': al_strategy, 
                        'Budget': budget, 
                        'Experiment': experiment, 
                        'Round': round_num,
                        'Statistic': stat_name, 
                        'Value': stat_value
                    }
                    data_records.append(record)
                # Add accuracy record
                data_records.append({
                    'AL Strategy': al_strategy, 
                    'Budget': budget, 
                    'Experiment': experiment, 
                    'Round': round_num,
                    'Statistic': 'Accuracy of hypothesized labels', 
                    'Value': accuracy
                })
            except IndexError:
                # Path format doesn't match expected pattern
                continue

    return pd.DataFrame(data_records)


# Example usage
# df = collect_data('/home/wassal/trust-wassal/tutorials/results/softloss0.3/pneumoniamnist/classimb/rounds10')





import plotly.graph_objs as go
from ipywidgets import widgets, HBox
from IPython.display import display

def interactive_visualization(df):
    # Get unique values for dropdowns
    experiments = df['Experiment'].unique()
    budgets = df['Budget'].unique()
    stats = df['Statistic'].unique()

    # Create dropdown widgets
    experiment_dropdown = widgets.Dropdown(options=experiments, description='Experiment:')
    budget_dropdown = widgets.Dropdown(options=budgets, description='Budget:')
    stat_dropdown = widgets.Dropdown(options=stats, description='Statistic:')

    # Plotting function
    def update_plot(*args):
        with fig.batch_update():
            fig.data = []
            experiment = experiment_dropdown.value
            budget = budget_dropdown.value
            stat = stat_dropdown.value
            filtered_df = df[(df['Experiment'] == experiment) & (df['Budget'] == budget) & (df['Statistic'] == stat)]

            for al_strategy in filtered_df['AL Strategy'].unique():
                strategy_df = filtered_df[filtered_df['AL Strategy'] == al_strategy]
                fig.add_trace(go.Scatter(x=strategy_df['Round'], y=strategy_df['Value'], 
                                         mode='lines+markers', name=al_strategy))

            fig.update_layout(title=f'{stat} for Experiment: {experiment}, Budget: {budget}',
                              xaxis_title='Round',
                              yaxis_title=stat,
                              legend_title='AL Strategy')

    # Initialize empty figure
    fig = go.FigureWidget()

    # Display widgets and initial plot
    experiment_dropdown.observe(update_plot, names='value')
    budget_dropdown.observe(update_plot, names='value')
    stat_dropdown.observe(update_plot, names='value')
    display(VBox([HBox([experiment_dropdown, budget_dropdown]), stat_dropdown]))
    update_plot()

    return fig
# Example usage
# df = collect_data('path_to_round10_folder')
# interactive_visualization(df)


# Example usage after collecting data
df = collect_data('/home/wassal/trust-wassal/tutorials/results/softloss0.3/pneumoniamnist/classimb/rounds10/')
interactive_visualization(df)
```
